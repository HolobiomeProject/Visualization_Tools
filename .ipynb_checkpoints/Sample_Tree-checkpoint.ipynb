```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
df = pd.read_csv('tree.csv',na_values=" NaN")
df = df.fillna('')
WstoMs = df.loc[df['Master Stock Link']!='']
WstoMs.index =  range(len(WstoMs))
MstoOG = df.loc[df['OG Link']!='']
MstoOG.index =  range(len(MstoOG))
OG = df.loc[(df['OG Link'] == '')&(df['Master Stock Link'] == '')]
OG.index =  range(len(OG))
```

```python
import matplotlib.pyplot as plt
import networkx as nx

#edges from Og to master stock
mtoo_edges = []
for i in range(0, len(OG['name'])):
    for j in range(0, len(MstoOG['OG Link'])):
        if OG['name'][i] == MstoOG['OG Link'][j]:
            mtoo_edges.append(tuple([OG['name'][i],MstoOG['name'][j]]))
#edges from Master stock to Working
wtoms_edges = []
for i in range(0, len(MstoOG['name'])):
    for j in range(0, len(WstoMs['Master Stock Link'])):
        if MstoOG['name'][i] == WstoMs['Master Stock Link'][j]:
            wtoms_edges.append(tuple([MstoOG['name'][i],WstoMs['name'][j]]))

#Absolute locaitons
#this can be cleaned up
pos = {}
y = 5
#og positions
for i in OG['name']:
    pos[i]=(0,y)
    y +=5
y = 5
#master stock to og
for k in MstoOG['name']:
    pos[k]=(5,y)
    y +=5  
#working stock to master
y = 5
for j in WstoMs['name']:
    pos[j]=(10,y)
    y +=5           
g=nx.Graph()
g.add_nodes_from(OG['name'])
g.add_nodes_from(WstoMs['name'])
g.add_nodes_from(MstoOG['name'])                    
g.add_edges_from(mtoo_edges)
g.add_edges_from(wtoms_edges)
nx.draw(g,pos = pos, with_labels = True,node_size=100,font_size=8, color = 'r')
```

```python
import plotly.plotly as py
import plotly.graph_objs as go

import jgraph
from jgraph import *
jgraph.__version__
```

```python
nr_vertices = 25
v_label = map(str, range(nr_vertices))
G = Graph.Tree(nr_vertices, 2) # 2 stands for children number
lay = G.layout('rt')

position = {k: lay[k] for k in range(nr_vertices)}
Y = [lay[k][1] for k in range(nr_vertices)]
M = max(Y)

es = EdgeSeq(G) # sequence of edges
E = [e.tuple for e in G.es] # list of edges

L = len(position)
Xn = [position[k][0] for k in range(L)]
Yn = [2*M-position[k][1] for k in range(L)]
Xe = []
Ye = []
for edge in E:
    Xe+=[position[edge[0]][0],position[edge[1]][0], None]
    Ye+=[2*M-position[edge[0]][1],2*M-position[edge[1]][1], None] 

labels = v_label
```

```python
lines = go.Scatter(x=Xe,
                   y=Ye,
                   mode='lines',
                   line=dict(color='rgb(210,210,210)', width=1),
                   hoverinfo='none'
                   )
dots = go.Scatter(x=Xn,
                  y=Yn,
                  mode='markers',
                  name='',
                  marker=dict(symbol='dot',
                                size=18, 
                                color='#6175c1',    #'#DB4551', 
                                line=dict(color='rgb(50,50,50)', width=1)
                                ),
                  text=labels,
                  hoverinfo='text',
                  opacity=0.8
                  )
```

```python
def make_annotations(pos, text, font_size=10, font_color='rgb(250,250,250)'):
    L=len(pos)
    if len(text)!=L:
        raise ValueError('The lists pos and text must have the same len')
    annotations = go.Annotations()
    for k in range(L):
        annotations.append(
            go.Annotation(
                text=labels[k], # or replace labels with a different list for the text within the circle  
                x=pos[k][0], y=2*M-position[k][1],
                xref='x1', yref='y1',
                font=dict(color=font_color, size=font_size),
                showarrow=False)
        )
    return annotations  
```

```python
axis = dict(showline=False, # hide axis line, grid, ticklabels and  title
            zeroline=False,
            showgrid=False,
            showticklabels=False,
            )

layout = dict(title= 'Tree with Reingold-Tilford Layout',  
              annotations=make_annotations(position, v_label),
              font=dict(size=12),
              showlegend=False,
              xaxis=go.XAxis(axis),
              yaxis=go.YAxis(axis),          
              margin=dict(l=40, r=40, b=85, t=100),
              hovermode='closest',
              plot_bgcolor='rgb(248,248,248)'          
              )
```

```python
data=go.Data([lines, dots])
fig=dict(data=data, layout=layout)
fig['layout'].update(annotations=make_annotations(position, v_label))
py.iplot(fig, filename='Tree-Reingold-Tilf')
```
