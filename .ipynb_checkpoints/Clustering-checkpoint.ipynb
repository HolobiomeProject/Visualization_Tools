# Cluster Map Use

Cluster maps by the seaborn library are very easy to work with and comes with an array of distance and linkage algorithms.

NOTE: Think about the distance and linkage options that best fit your data.

```python
import seaborn as sns; sns.set(color_codes=True)
import pandas as pd
```

```python
bugs = pd.read_csv('DataExamples/ClusterExample.csv')
species = bugs.pop("Species")
```

```python
g = sns.clustermap( bugs, 
                    row_cluster = True,
                    col_cluster = True,
                    metric = 'correlation',#distance
                    method = 'average',#linkage
                    robust = True,
                    z_score = True,
                    cmap = 'coolwarm',
                    yticklabels = species )
```
